```python
import yfinance as yf
import pandas as pd
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
```

```python
tesla_data = yf.download('TSLA', start='2022-01-01', end='2022-12-31')
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
tesla_revenue_url = 'https://ycharts.com/companies/TSLA/revenue'
tesla_revenue_html = requests.get(tesla_revenue_url).text
tesla_revenue_soup = BeautifulSoup(tesla_revenue_html, 'html.parser')
```

```python
tesla_revenue = pd.DataFrame(tesla_revenue_soup)
print(tesla_revenue.tail())
```

```python
gme_data = yf.download('GME', start='2022-01-01', end='2022-12-31')
gme_data.reset_index(inplace=True)
print(gme_data.head())
```

```python
gme_revenue_url = 'https://ycharts.com/companies/GME/revenue'
gme_revenue_html = requests.get(gme_revenue_url).text
gme_revenue_soup = BeautifulSoup(gme_revenue_html, 'html.parser')
# Parse the data and create the dataframe
gme_revenue = pd.DataFrame(gme_revenue_soup)  # Use appropriate code to extract data
print(gme_revenue.tail())
```

```python
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.show()

make_graph(tesla_data, 'Tesla Stock Data')
```

```python
make_graph(gme_data, 'GameStop Stock Data')
```
